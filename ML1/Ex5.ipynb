```python
a = pr.read_mat("triclust")
pr.fusion_graph(+a, 'single')
```

## b

```python
pr.fusion_graph(+a, 'complete')
```

 The fusion graph is ambiguous about the exact number of clusters: either two or three clustersare associated with large fusion jumps of roughly the same magnitude. This is due to the completelinkage distance: distance between furthest samples of clusters at (0,0) and (1,1) (which aremerged to result from a total of two clusters) is roughly half the distance between the distancebetween the furthest samples in the cluster at (0,0) and (2,2), which are merged to result ina single cluster. In single linkage this does not occur, since the minimal distances between theclusters at (0,0) and (1,1) is the same as the minimal distance between the clusters at (1,1) and(2,2)
